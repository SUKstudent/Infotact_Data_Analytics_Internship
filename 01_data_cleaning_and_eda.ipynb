Clean the Dataset

```python
df["event_timestamp_utc"] = pd.to_datetime(df["event_timestamp_utc"])
df["event_date"] = pd.to_datetime(df["event_date"])
df["conversion_timestamp_utc"] = pd.to_datetime(df["conversion_timestamp_utc"], errors="coerce")

df["ad_spend"] = pd.to_numeric(df["ad_spend"], errors="coerce").fillna(0)
df["conversion_value"] = pd.to_numeric(df["conversion_value"], errors="coerce").fillna(0)

df = df.drop_duplicates(subset=["event_id"])

df = df.sort_values(["journey_id", "event_timestamp_utc"])

df.head()
```

```python
df.to_csv("cleaned_mta_roi_dataset.csv", index=False)
```

Calculate Main Numbers

```python
total_spend = df["ad_spend"].sum()
total_revenue = df["conversion_value"].sum()
total_conversions = df["is_conversion"].sum()
total_journeys = df["journey_id"].nunique()

roas = total_revenue / total_spend
cac = total_spend / total_conversions
conversion_rate = total_conversions / total_journeys

print("Total Spend:", round(total_spend, 2))
print("Total Revenue:", round(total_revenue, 2))
print("Total Conversions:", total_conversions)
print("Total Journeys:", total_journeys)
print("ROAS:", round(roas, 2))
print("CAC:", round(cac, 2))
print("Conversion Rate:", round(conversion_rate * 100, 2), "%")
```

Channel-wise Analysis

```python
channel_summary = df.groupby("channel").agg(
    total_spend=("ad_spend", "sum"),
    total_revenue=("conversion_value", "sum"),
    conversions=("is_conversion", "sum"),
    touchpoints=("event_id", "count")
).reset_index()

channel_summary["roas"] = channel_summary["total_revenue"] / channel_summary["total_spend"]
channel_summary["cac"] = channel_summary["total_spend"] / channel_summary["conversions"]

channel_summary = channel_summary.replace([float("inf"), -float("inf")], 0)
channel_summary = channel_summary.fillna(0)

channel_summary
```

Save Channel Summary

```python
channel_summary.to_csv("channel_summary.csv", index=False)
```
